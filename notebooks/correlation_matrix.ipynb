```python
# Import libraries
import pandas as pd
import seaborn as sns
```

```python
# Import dataset
df19 = pd.read_csv("../datasets/data2019.csv")
df20 = pd.read_csv("../datasets/data2020.csv")
# Drop columns that are not dependent nor independent variables
df19 = df19.iloc[:,6:]
df20 = df20.iloc[:,5:]

# Create a matrix that contains correlation coefficient
cor19 = df19.corr()
cor20 = df20.corr()
```

```python
# Create a correlation matrix of year 2019 data
sns.heatmap(cor19, vmin=-1, vmax=1, mask=pd.np.tril(cor19), annot=True)
```

```python
# Create a correlation matrix of year 2020 data
sns.heatmap(cor20, vmin=-1, vmax=1, mask=pd.np.tril(cor20), annot=True)
```

```python
# Create a correlation matrix of 2020 vs 2019 data
corr_chg = cor20 - cor19
sns.heatmap(corr_chg.mask(abs(corr_chg) < 0.1), vmin=-1, vmax=1, mask = pd.np.tril(corr_chg), annot=True)
```
